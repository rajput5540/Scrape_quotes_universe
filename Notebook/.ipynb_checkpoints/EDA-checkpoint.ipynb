```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
import warnings
warnings.filterwarnings("ignore")
```

# Basic EDA

## Q1 : Load the CSV into a Pandas DataFrame

```python
quotes_df = pd.read_csv("Quotes.csv")
```

```python
print("First 5 rows of the dataset:")
quotes_df.head()
```

## Q2 : Display basic statistics about the dataset (e.g., missing values, unique authors, number of tags, etc.)

```python
print("\nDataset Info:")
quotes_df.info()
```

```python
print("\nNumber of unique authors:", quotes_df['author'].nunique())
print("Number of quotes:", len(quotes_df))
print("\nMissing values")
print(quotes_df.isnull().sum())
```

# Data Visualization

## Q1 : Bar chart of the top 10 most quoted authors.

```python
author_counts = quotes_df['author'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=author_counts.values, y=author_counts.index, palette='cubehelix')
plt.title("Top 10 Most Quoted Authors")
plt.xlabel("Number of Quotes")
plt.ylabel("Author")
plt.show()
```

## Q2 : Word cloud of the most common words in the quotes.

```python
all_quotes = " ".join(quotes_df['quote'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_quotes)
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Words in Quotes")
plt.show()
```

## Q3 : Pie chart showing the proportion of top 5 tags.

```python
tag_list = ",".join(quotes_df['tag_name'].dropna().astype(str)).split(",")
tag_counts = Counter([tag.strip() for tag in tag_list if tag.strip()])
top_tags = tag_counts.most_common(5)
labels, sizes = zip(*top_tags)

# Colors (you can choose any palette you like)
colors = ['#66b3ff', '#ff9999', '#99ff99', '#ffcc99', '#c2c2f0']

# Exploding the first slice a little
explode = (0.1, 0.05, 0.05, 0.05, 0.05)

# Plot
plt.figure(figsize=(10, 8))
plt.pie(
    sizes, 
    labels=labels, 
    colors=colors,
    autopct='%1.2f%%',  # prettier percentages
    startangle=140,
    explode=explode,    # separate the slices
    shadow=True,        # add shadow for 3D look
    textprops={'fontsize': 12}  # control font size
)
plt.title("Top 5 Tags (Proportion)", fontsize=16)
plt.axis('equal')  # equal aspect ratio ensures pie is a circle
plt.show()
```

## Q4 : Distribution of number of tags per quote

```python
quotes_df['tag_count'] = quotes_df['tag_name'].fillna('').apply(lambda x: len(str(x).split(',')))
plt.figure(figsize=(8, 6))
sns.histplot(quotes_df['tag_count'], bins=range(1, 8), kde=False)
plt.title("Distribution of Number of Tags per Quote")
plt.xlabel("Number of Tags")
plt.ylabel("Number of Quotes")
plt.xticks(range(1, 8))
plt.tight_layout()
plt.show()
```

## Q5 : Average Quote Length per Author

```python
quotes_df['quote_length'] = quotes_df['quote'].str.len()
avg_length = quotes_df.groupby('author')['quote_length'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_length.values, y=avg_length.index, palette='viridis')
plt.title("Average Quote Length per Author")
plt.xlabel("Average Length")
plt.grid(False)
plt.show()
```

## Q6 : Distribution of Quote Lengths

```python
plt.figure(figsize=(10, 6))
sns.histplot(quotes_df['quote_length'], bins=30, kde=True, color= 'purple')
plt.gca().lines[0].set_color('orange')  # Change the KDE line color to orange
plt.title("Distribution of Quote Lengths")
plt.xlabel("Length of Quote")
plt.ylabel("Frequency")
plt.grid(False)
plt.show()
```

## Q7 : Most Common Starting Words in Quotes

```python
quotes_df['first_word'] = quotes_df['quote'].str.extract(r'^\W*(\w+)', expand=False)
top_first_words = quotes_df['first_word'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_first_words.index, y=top_first_words.values, palette='magma')
plt.title("Most Common First Words in Quotes")
plt.ylabel("Frequency")
plt.xlabel("First Word")
plt.grid(axis = 'y', linestyle='--', alpha=0.5)
plt.show()
```
